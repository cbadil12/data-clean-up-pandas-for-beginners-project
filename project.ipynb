#### Exercise 01. Basic data analysis with native Python structures.  
Create a list with the grades of 5 students: [8.5, 9.2, 7.8, 8.9, 10].

- Calculate the average of the grades.

```python
#Method without libraries
#--------------------------

#Set data with grades
grades=[8.5,9.2,7.8,8.9,10]

#Aux variable to compute total
total=0

#Loop to acumulate grades
for i in grades:
    total = total + i

#Calculate average
ave_grade=total/len(grades)

print(ave_grade)
```

- Find the highest and lowest grade.

```python
#Initialize values
min_grade=grades[0]
max_grade=grades[0]

#Loop to find min and max
for i in grades:
    if i < min_grade:
        min_grade=i
    if i > max_grade:
        max_grade=i
        
print(min_grade,max_grade)
```

#### Exercise 01. What is the most expensive house in the entire dataset? (‚òÖ‚òÜ‚òÜ)

Print the address and price of the selected house. For example:

`The house located at Calle del Prado, N¬∫20 is the most expensive, and its price is 5000000 USD.`

```python
#Find the most expensive price
max_price=ds["price"].max()

#Find the index correspoding to that price
max_index=ds.index[ds['price'] == max_price]

#Find the address correspoding to that index
max_address=ds["address"].iloc[max_index]

#Final print
print("The house located at " + max_address + " is the most expensive, and its price is " + str(max_price)+" USD")
```

#### Exercise 02. What is the cheapest house in the dataset? (‚òÖ‚òÜ‚òÜ)

This exercise is similar to the previous one, except now we are looking for the house with the lowest price. Remember to print the address and price of the selected house. For example:

`The house located at Calle Alcal√°, N¬∫58 is the cheapest, and its price is 12000 USD.`


```python
#Method A) printing all min (price = 0)
#--------------------------

#Find the cheapest price
min_price=ds["price"].min()

#Find the index correspoding to that price
min_index=ds.index[ds['price'] == min_price]

#Find the address correspoding to that index
min_address=ds["address"].iloc[min_index]

#Final print
print("The house located at " + min_address.iloc[0] + " is the cheapest, and its price is " + str(min_price)+" USD")


#--------------------------
#Method B) eluding prices = 0
#--------------------------

#Clean rows with price equal to 0
ds_clean=ds.loc[ds["price"]!=0]

#Find the cheapest price
min_price_clean=ds_clean["price"].min()

#Find the index correspoding to that price
min_index_clean=ds.index[ds['price'] == min_price_clean]

#Find the address correspoding to that index
min_address_clean=ds["address"].iloc[min_index_clean]

#Final print
print("The house located at " + min_address_clean.iloc[0] + " is the cheapest, and its price is " + str(min_price_clean)+" USD")
```

#### Exercise 07. What is the average price in the population of "Arroyomolinos (Madrid)"? (‚òÖ‚òÖ‚òÜ)

Print the value obtained from the 'level5' column.

```python
ds_aux=ds[ds["level5"]=="Arroyomolinos (Madrid)"]
mean_price=ds_aux["price"].mean()
print(mean_price)
```

#### Exercise 08. Plot the histogram of prices for the population of "Arroyomolinos (Madrid)" and explain what you observe (‚òÖ‚òÖ‚òÜ)

Print the histogram of the prices and write a brief analysis of the plot in the Markdown cell.

```python
import matplotlib.pyplot as plt

#Calculate auxs
mean_price=ds_aux["price"].mean()
median_price=ds_aux["price"].median()
size=ds_aux["price"].size
variance=ds_aux["price"].var()
sigma=variance**0.5

diff_price=round((mean_price-median_price)/mean_price,3)
print(diff_price)

#Set the figure
plt.figure(figsize = (10, 5), facecolor="w", edgecolor="grey")
plt.hist(ds_aux["price"],bins=60, color="cyan", edgecolor="black")
plt.title("Arroyomolinos (Madrid)")
plt.xlabel("Price [USD]")
plt.ylabel("Properties [uds]")

plt.axvline(mean_price, linestyle="--", linewidth=2, color="red", label="Mean price: "+ str(round(mean_price,2)) + " USD")
plt.axvline(median_price, linestyle="--", linewidth=2, color="blue", label="Median price: "+ str(round(median_price,2)) + " USD")
plt.axvspan(mean_price - sigma, mean_price + sigma, alpha=0.25, label="Standard Deviation from mean")

plt.text(0, 16,
    "BRIEF ANALYSIS:" +
    "\n- No. properties analyzed: " + str(size) +
    "\n- Mean and median price differ by " + str(100*diff_price) + "%",
    va="top", ha="left",fontsize=9,
    bbox=dict(boxstyle="round", facecolor="cyan", edgecolor="black", alpha=0.25)
)

plt.legend()

#ax.axvline(mediana, ls=":",  lw=1)
plt.show()
```
